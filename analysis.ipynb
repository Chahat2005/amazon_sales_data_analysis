```python
data['discounted_price']=data['discounted_price'].str.replace('₹','')
```

```python
data['discounted_price']=data['discounted_price'].str.replace(',','')
```

```python
data['actual_price']=data['actual_price'].str.replace('₹','')
data['actual_price']=data['actual_price'].str.replace(',','')
data['actual_price']=data['actual_price'].astype(float)
```

```python
data['discount_percentage']=data['discount_percentage'].str.replace('%','')
data['discount_percentage']=data['discount_percentage'].astype(float)
```

```python
data['rating'].unique()
```

```python
data['rating']=data['rating'].replace("|","0")
```

```python
data['rating']=data['rating'].astype(float)
```

```python
data['rating_count']=data['rating_count'].str.replace(',','')
data['rating_count']=data['rating_count'].astype(float)
```

```python
plt.figure(figsize=(20,15))

category = ['Main_Category', 'rating']

for i in range(0,len(category)):
    plt.subplot(2,2,i+1)
    sns.countplot(x=data[category[i]], palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Insights:
The dataset is dominated by a few main categories, showing that Amazon focuses heavily on selected product segments.
Product ratings are mostly concentrated around 4 and above, indicating generally positive customer satisfaction.

Most Popular Category

```python
import matplotlib.pyplot as plt

top_cat = data['category'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_cat.plot(kind='bar')
plt.title("Top 10 Most Popular Categories")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=45)
plt.show()
```

Insights:
A few categories dominate the dataset, with the top categories having significantly higher product counts than others. This indicates strong product availability and demand in these categories.

2. Highest Reviews by Category

```python
top_reviews = data.groupby('category')['rating_count'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_reviews.plot(kind='bar', color='orange')
plt.title("Top Categories by Total Reviews")
plt.xlabel("Category")
plt.ylabel("Total Rating Count")
plt.xticks(rotation=45)
plt.show()
```

Insights:Some categories receive significantly higher total reviews than others, indicating higher customer engagement and popularity in these categories.

3. Highest Average Discount Category

```python
avg_discount = data.groupby('category')['discount_percentage'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
avg_discount.plot(kind='bar', color='green')
plt.title("Top Categories by Average Discount")
plt.xlabel("Category")
plt.ylabel("Average Discount %")
plt.xticks(rotation=45)
plt.show()
```

Insights:Some categories offer significantly higher average discounts than others, indicating strong pricing competition in these segments.

4. Discount vs Rating (Scatter Plot)

```python
plt.figure(figsize=(8,5))

plt.scatter(data['discount_percentage'], data['rating'])

plt.title("Discount vs Rating")
plt.xlabel("Discount Percentage")
plt.ylabel("Rating")
plt.show()
```

5. Best Value for Money Products

```python
best_products = data[(data['rating'] >= 4.3) & (data['discount_percentage'] > 30)] \
                .sort_values('rating_count', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.barh(best_products['product_name'], best_products['rating_count'], color='purple')

plt.title("Best Value for Money Products")
plt.xlabel("Rating Count")
plt.ylabel("Product Name")
plt.gca().invert_yaxis()
plt.show()
```
